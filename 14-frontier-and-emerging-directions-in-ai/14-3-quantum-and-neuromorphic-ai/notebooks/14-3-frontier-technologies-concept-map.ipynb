## Concept Map: Mainstream AI -> Frontier Hardware AI

### Mainstream AI Stack
- Data pipelines -> model training -> deployment -> monitoring.
- Hardware baseline: CPU/GPU/TPU-centric workflows.

### Quantum ML Branch
- Core ideas: qubits, superposition, entanglement, interference.
- Typical framing: hybrid quantum-classical algorithms.
- Candidate domains: optimization, chemistry simulation, specific kernels.

### Neuromorphic Branch
- Core ideas: event-driven spikes, co-located memory and compute.
- Typical framing: ultra-low-power edge inference and sensing.
- Candidate domains: robotics, IoT, always-on anomaly detection.

### Cross-Cutting Decisions
- Problem fit over novelty.
- Benchmark against strong classical baselines.
- Pilot first, scale only after evidence.

## Practical Learning Paths

### Awareness Path (4-6 weeks)
- Read one high-quality overview per week.
- Track one benchmark-driven case per domain.
- Summarize where claims are speculative vs validated.

### Applied Engineer Path (2-3 months)
- Run one toy pilot with clear baseline comparison.
- Document assumptions, metrics, and failure modes.
- Present go/no-go recommendation.

### Specialist Collaboration Path
- Partner with hardware or domain researchers.
- Scope narrow prototypes with explicit handoff boundaries.

### Exceptions

## Interview Questions & Answers

1. **Q:** What is quantum ML in simple terms?  
   **A:** A set of hybrid algorithms exploring whether quantum state operations improve specific ML tasks.
2. **Q:** What is neuromorphic computing?  
   **A:** Brain-inspired hardware/algorithms focused on event-driven, energy-efficient computation.
3. **Q:** Why not migrate all models to frontier hardware?  
   **A:** Tooling and practical advantages are still domain-specific and often limited.
4. **Q:** What is a qubit intuition?  
   **A:** A quantum unit that can encode superposed states until measurement.
5. **Q:** SNN advantage at edge?  
   **A:** Sparse event-driven updates can reduce power usage.
6. **Q:** How do you evaluate hype vs value?  
   **A:** Benchmark against strong classical baselines under realistic constraints.
7. **Q:** What role should an applied AI engineer play?  
   **A:** Be frontier-aware, run focused pilots, and collaborate with specialists.
8. **Q:** What is a safe pilot strategy?  
   **A:** Narrow scope, fixed metrics, clear exit criteria.
9. **Q:** One red flag in frontier proposals?  
   **A:** No baseline comparison or unclear problem fit.
10. **Q:** When is neuromorphic most compelling?  
    **A:** Strict low-power, always-on sensing scenarios.
11. **Q:** When is quantum ML most relevant?  
    **A:** Specialized optimization/simulation domains where quantum advantage is plausible.
