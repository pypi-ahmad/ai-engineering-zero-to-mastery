For each case:

4. Harm pathways and mitigation options.
5. Decision recommendation and residual risk.


## Case Study 1: Algorithmic Hiring

### Scenario

A company uses an AI classifier to shortlist candidates from resumes.

### Stakeholders

- Applicants
- Recruiters/HR
- Legal/compliance teams
- Organization leadership

### Ethical Issues

- Potential proxy discrimination
- Opaque rejection decisions
- Reduced human review for borderline cases

### Mitigation Ideas

- Structured appeal process
- Human review for uncertain or high-impact decisions
- Group-level fairness monitoring


## Case Study 2: Predictive Policing / Risk Scoring

### Scenario

A city uses historical arrest data to allocate patrol resources.

### Ethical Issues

- Historical data may encode structural bias
- Feedback loops reinforce over-policing
- Potential chilling effects on civil liberties

### Practical Response

- Independent impact assessment
- Continuous disparity monitoring
- Public oversight and contestability mechanisms


## Case Study 3: Social Recommender Systems

### Scenario

A platform optimizes ranking for engagement and watch time.

### Ethical Issues

- Manipulation of attention and autonomy
- Amplification of misinformation/extreme content
- Disproportionate effects on younger users

### Mitigation Directions

- Multi-objective optimization beyond engagement
- Friction for harmful content propagation
- Transparency controls and user choice features


## Optional Quantitative Lens: Group Fairness Snapshot

Below is a toy example to compute acceptance rates by group. This is **not** a full fairness audit; it is a first diagnostic.


```python
df = pd.DataFrame(
    {
        "group": ["A", "A", "A", "B", "B", "B", "B"],
        "approved": [1, 0, 1, 1, 0, 0, 0],
    }
)

acceptance = df.groupby("group")["approved"].mean().rename("acceptance_rate")
acceptance
```

```python
reference = acceptance.max()
parity_ratio = (acceptance / reference).rename("selection_rate_ratio_vs_best_group")
pd.concat([acceptance, parity_ratio], axis=1)
```

## Connect to Theory

- Consequentialist analysis asks whether net societal benefit is positive.
- Deontological analysis asks whether rights and duties are respected.
- Virtue ethics asks whether institutions are acting with responsibility and integrity.

In practice, robust ethical analysis uses all three lenses because each catches different failure modes.


## Business Case Studies & Exceptions

### Case A: Recommender Fairness vs Growth

A media company found that boosting engagement increased ad revenue but worsened exposure inequality for minority creators. The final policy introduced fairness constraints and creator diversity safeguards despite a short-term KPI dip.

### Case B: Hiring Model in Regulated Region

A multinational delayed rollout of an automated hiring model because explainability and appeal mechanisms were insufficient for local legal expectations. A phased deployment with mandatory human review reduced legal risk.

### Exceptions

- In low-impact internal ranking tools, lightweight governance may be proportionate.
- In hiring, credit, health, justice, and education, strong controls are non-negotiable.


## Interview Questions & Answers

1. **Q: What is algorithmic bias?**
   **A:** Systematic unfair outcomes caused by data, model, or deployment context.
2. **Q: Why is hiring AI ethically sensitive?**
   **A:** It affects opportunity, livelihood, and protected rights.
3. **Q: How do you evaluate fairness quickly?**
   **A:** Start with subgroup outcome/error rates, then investigate causal drivers.
4. **Q: Why isn't high accuracy enough?**
   **A:** Aggregate accuracy can hide severe harms for specific groups.
5. **Q: What is contestability?**
   **A:** The ability for affected people to challenge and review decisions.
6. **Q: How can recommenders harm autonomy?**
   **A:** By using manipulative ranking dynamics that shape behavior without informed consent.
7. **Q: What is a feedback loop in predictive policing?**
   **A:** Deployment generates new biased data that reinforces the model's original bias.
8. **Q: When should humans stay in the loop?**
   **A:** For high-impact or low-confidence decisions.
9. **Q: How do ethics and compliance differ?**
   **A:** Compliance is legal minimum; ethics includes broader responsibility and societal impact.
10. **Q: What is one practical mitigation for hiring AI?**
   **A:** Add structured human review + appeal process + subgroup monitoring.
