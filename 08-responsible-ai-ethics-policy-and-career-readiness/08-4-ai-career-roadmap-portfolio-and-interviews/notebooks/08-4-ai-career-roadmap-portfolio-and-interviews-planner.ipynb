# Lesson 8.4 - AI Career Roadmap, Portfolio & Interviews Planner

## Objectives

- Build a personalized 6-12 month learning plan.
- Convert lessons into portfolio artifacts.
- Prepare technical and behavioral interview stories.


## Skill Gap Analysis Template

Rate each area from 1 (beginner) to 5 (advanced):

- Python and software engineering
- Math and statistics
- Classical ML and evaluation
- Deep learning and transformers
- RAG and agentic workflows
- MLOps/LLMOps and monitoring
- Responsible AI and governance
- Product and communication


```python
from __future__ import annotations

import json
from dataclasses import dataclass, asdict
from typing import List, Dict
```

## Planner Helper (Generate Roadmap JSON)


```python
@dataclass
class TrackPreference:
    role_target: str
    months: int
    weekly_hours: int
    focus_areas: List[str]


def generate_roadmap(pref: TrackPreference) -> Dict[str, object]:
    phases = [
        {"phase": "Foundations", "duration_months": 2, "outcome": "Strong coding/math + baseline ML project"},
        {"phase": "Modeling", "duration_months": 2, "outcome": "Classical ML + robust evaluation"},
        {"phase": "GenAI", "duration_months": 2, "outcome": "RAG/LLM app with eval"},
        {"phase": "Production", "duration_months": 2, "outcome": "Deployment, monitoring, governance"},
        {"phase": "Portfolio & Interviews", "duration_months": 2, "outcome": "Top projects polished + interview readiness"},
    ]
    return {
        "target_role": pref.role_target,
        "timeline_months": pref.months,
        "weekly_hours": pref.weekly_hours,
        "focus_areas": pref.focus_areas,
        "phases": phases,
    }


plan = generate_roadmap(
    TrackPreference(
        role_target="GenAI Engineer",
        months=10,
        weekly_hours=12,
        focus_areas=["RAG", "evaluation", "agentic workflows", "LLMOps"],
    )
)
print(json.dumps(plan, indent=2))
```

## Project Planning Sheet Template

For each portfolio project:

- Problem and user
- Why this matters (business KPI)
- Data / tools / model stack
- Evaluation plan (offline + online proxies)
- Deployment and monitoring plan
- Responsible AI risks and mitigations
- Demo artifact (repo + video + one-page case summary)


## 6-12 Month Execution Cadence

- Weekly: build + review + publish notes.
- Monthly: complete one milestone deliverable.
- Quarterly: ship one interview-ready project artifact.


## Business Case Studies & Exceptions

### Case A: Student Profile

A final-year student used a 9-month plan, shipped three projects (ML, RAG, deployment), and added concise write-ups with metrics and failure analysis. Recruiter callbacks improved because evidence was concrete.

### Case B: Working Professional Transition

A backend developer allocated 10 hours/week for 10 months, reused prior software strengths, and shifted into applied AI platform roles by showcasing reliability-focused projects.

### Exceptions

- If targeting research scientist roles, prioritize papers/theory depth over broad product portfolio.
- If already senior in MLOps, focus portfolio on GenAI/RAG gaps rather than redoing fundamentals.


## Interview Questions & Answers

1. **How do you choose target AI role?**  
   Match strengths, interests, and market demand, then align projects to that role.
2. **What should your first portfolio project prove?**  
   End-to-end execution, not only model training.
3. **How many projects should be interview-ready?**  
   At least 3 high-quality projects with clear storytelling.
4. **How do you discuss project impact without production users?**  
   Use baseline comparisons and realistic proxy KPIs.
5. **What is a strong STAR answer for AI interviews?**  
   One showing technical depth, trade-offs, and measurable results.
6. **How should career changers position themselves?**  
   Emphasize transferable skills plus evidence of AI delivery.
7. **How do you avoid tutorial-style portfolio weakness?**  
   Add original problem framing and rigorous evaluation choices.
8. **How do you prepare for system design rounds?**  
   Practice full-stack AI architecture including reliability and governance.
9. **How do you present Responsible AI work?**  
   Show risk assessment, controls, and concrete mitigation outcomes.
10. **What is one hiring red flag?**  
   Inability to explain your own design decisions and trade-offs.
