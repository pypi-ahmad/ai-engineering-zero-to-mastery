## Mapping System Types to Regulatory Concerns

| System Type | Typical Risk Signals | Governance Needs |
|---|---|---|
| Hiring / credit / education ranking | Fundamental-rights impact, discrimination risk | Fairness testing, explainability, appeals, audits |
| Biometric identification | High privacy/surveillance risk | Strict legal basis, access controls, impact assessment |
| Low-risk internal assistant | Limited external impact | Baseline policy, logging, usage constraints |
| Public-sector decision support | Accountability and public trust stakes | Transparency, human oversight, stronger records |


## Template: Model/System Card Fields

- Intended purpose and out-of-scope use
- Data sources and lineage
- Performance metrics and subgroup metrics
- Known limitations and failure modes
- Human oversight requirements
- Governance owner and contact


## Template: AI Impact Assessment (AIIA) Fields

1. System description
2. Stakeholders and affected populations
3. Risk categories and severity/likelihood
4. Mitigation controls and residual risk
5. Monitoring plan and escalation owners


```python
from __future__ import annotations

REQUIRED_METADATA = {
    "model_name",
    "version",
    "intended_use",
    "risk_tier",
    "data_lineage",
    "owner",
    "last_review_date",
}


def governance_gate(metadata: dict) -> tuple[bool, list[str]]:
    missing = sorted(list(REQUIRED_METADATA - set(metadata.keys())))
    if missing:
        return False, missing
    if metadata.get("risk_tier") == "high" and not metadata.get("human_oversight"):
        return False, ["human_oversight is required for high-risk systems"]
    return True, []


example_metadata = {
    "model_name": "credit-risk-v4",
    "version": "4.0.0",
    "intended_use": "loan pre-screening",
    "risk_tier": "high",
    "data_lineage": "warehouse.credit_snapshots_2026q1",
    "owner": "risk-ml-team",
    "last_review_date": "2026-06-20",
    "human_oversight": True,
}

governance_gate(example_metadata)
```

## Connect to Theory

- Risk tiers influence control intensity.
- Governance turns abstract accountability into concrete artifacts and ownership.
- Legal alignment is operational: metadata completeness, review evidence, and deployment constraints.


## Business Case Studies & Exceptions

### Case A: Credit Risk Under Strict Regulation

A lender embedded metadata gates and adverse-action explanation checks in deployment pipelines. This reduced emergency rollbacks during audits.

### Case B: AI Tutor in Education

An edtech company enforced transparency notices and teacher override for high-impact recommendations, reducing complaints and governance risk.

### Exceptions

- Early prototypes may use lightweight governance, but production systems need full documentation.
- Cross-border products may require region-specific controls and data handling policies.


## Interview Questions & Answers

1. **What is the risk-based idea behind AI regulation?**  
   Higher potential harm requires stronger obligations and oversight.
2. **Why is governance an engineering responsibility?**  
   Compliance depends on system design, logs, and deployment controls.
3. **What is an AI impact assessment?**  
   Structured analysis of harms, mitigations, and residual risk.
4. **How do engineers help legal teams?**  
   By providing traceable artifacts and implementing enforceable controls.
5. **Why are model cards useful?**  
   They communicate intended use, limits, and accountability details.
6. **What is a deployment governance gate?**  
   A rule that blocks release when required metadata/approvals are missing.
7. **What is accountability in production AI?**  
   Named owners for performance, incidents, and remediation.
8. **How do you support auditability?**  
   Preserve immutable decision logs and change history.
9. **What is a common policy-vs-engineering gap?**  
   Policies exist, but controls are not encoded in delivery pipelines.
10. **How do you reduce legal risk in high-impact systems?**  
   Combine robust documentation, human oversight, and continuous monitoring.
