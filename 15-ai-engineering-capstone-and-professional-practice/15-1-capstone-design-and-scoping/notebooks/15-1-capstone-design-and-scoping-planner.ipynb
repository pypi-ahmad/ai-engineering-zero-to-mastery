# Lesson 15.1 - Capstone Design & Scoping (planner notebook)

This notebook helps you select and scope a capstone with explicit feasibility, impact, and delivery constraints.
    


## Objectives

- Build a project charter and statement of work (SoW).
- Score candidate capstone ideas with a repeatable framework.
- Select an MVP scope that is realistic and portfolio-strong.
    


## Project Charter Template

- **Background**
- **Problem statement**
- **Primary objective**
- **Scope in / out**
- **Users / stakeholders**
- **Timeline and milestones**
- **Top risks and mitigations**
- **Deliverables**
    


## Statement of Work (SoW) Template

1. Project title
2. Objective and non-objectives
3. Deliverables + acceptance criteria
4. Milestones + deadlines
5. Dependencies and assumptions
6. Risk register + contingencies
7. Communication cadence
    


```python
import pandas as pd

ideas = pd.DataFrame(
    [
        {"idea": "Customer support RAG assistant", "impact": 5, "feasibility": 4, "data_availability": 4, "novelty": 3, "career_relevance": 5},
        {"idea": "Medical imaging classifier benchmark", "impact": 4, "feasibility": 3, "data_availability": 2, "novelty": 4, "career_relevance": 4},
        {"idea": "Fraud detection ML pipeline + API", "impact": 5, "feasibility": 5, "data_availability": 4, "novelty": 3, "career_relevance": 4},
        {"idea": "Agentic travel planner", "impact": 3, "feasibility": 2, "data_availability": 3, "novelty": 5, "career_relevance": 4},
    ]
)
ideas
    
```

## Idea Scoring Helper


```python
def score_idea(row, w_impact=0.30, w_feasibility=0.30, w_data=0.20, w_novelty=0.10, w_career=0.10):
    return (
        w_impact * row["impact"]
        + w_feasibility * row["feasibility"]
        + w_data * row["data_availability"]
        + w_novelty * row["novelty"]
        + w_career * row["career_relevance"]
    )

ideas["score"] = ideas.apply(score_idea, axis=1)
ideas.sort_values("score", ascending=False)
    
```

```python
def classify_scope(score: float) -> str:
    if score >= 4.4:
        return "Strong candidate"
    if score >= 3.8:
        return "Feasible with strict scope"
    return "High risk / revise"

ideas["scope_assessment"] = ideas["score"].apply(classify_scope)
ideas[["idea", "score", "scope_assessment"]].sort_values("score", ascending=False)
    
```

## Scoping Checklist

- [ ] Problem statement is testable.
- [ ] Primary metric and threshold are explicit.
- [ ] Baseline approach is defined.
- [ ] Data source and fallback source are identified.
- [ ] Must/Should/Could scope tiers are documented.
- [ ] Milestone dates have owner and acceptance criteria.
- [ ] Top risks and mitigations are listed.
    


## Capstone & Practice Case Studies & Exceptions

### Case Study A: High-Scoring, Low-Risk Choice
`Fraud detection ML pipeline + API` scored high on feasibility and impact, making it suitable for a complete end-to-end capstone.

### Case Study B: High Novelty, Low Feasibility
`Agentic travel planner` had novelty but weak feasibility under timeline constraints; better as a stretch extension.

### Exception Pattern
If your data source is uncertain, prioritize a narrower problem with guaranteed data access.
    


## Interview Questions & Answers

1. **Q:** How did you choose your capstone topic?  
   **A:** I scored ideas on impact, feasibility, data readiness, novelty, and career fit.
2. **Q:** Why use a project charter?  
   **A:** It aligns team and stakeholders on objective, scope, and constraints.
3. **Q:** What is a realistic capstone scope?  
   **A:** One complete workflow with measurable outcomes and clear deliverables.
4. **Q:** How do you avoid over-scoping?  
   **A:** Use must/should/could tiers and milestone gates.
5. **Q:** What are essential success criteria?  
   **A:** Metric target, delivery completeness, and reproducibility.
6. **Q:** Why include non-objectives in SoW?  
   **A:** To prevent scope creep and reduce ambiguity.
7. **Q:** What if impact is high but data is weak?  
   **A:** Re-scope to a feasible variant or pick another idea.
8. **Q:** How do you make a capstone interview-ready?  
   **A:** Ensure end-to-end narrative with evidence-backed decisions.
9. **Q:** What is your week-1 deliverable?  
   **A:** Finalized charter, SoW, and milestone/risk plan.
10. **Q:** What triggers a project pivot?  
    **A:** Failed data assumptions, infeasible compute, or low stakeholder value.
11. **Q:** How did you balance novelty and feasibility?  
    **A:** Novel features were deferred until MVP completion.

## Capstone Scoring Rubric Worksheet
Score each dimension 1-5.

- Problem framing and scope control
- Architecture and implementation quality
- Evaluation rigor and reliability evidence
- Deployment and operations readiness
- Documentation and reproducibility
- Communication and reflection


```python
rubric_scores = {
    "problem_scope": 4,
    "architecture_impl": 4,
    "evaluation_rigor": 3,
    "ops_readiness": 3,
    "documentation": 4,
    "communication_reflection": 4,
}

weighted = {
    "problem_scope": 0.20,
    "architecture_impl": 0.25,
    "evaluation_rigor": 0.20,
    "ops_readiness": 0.15,
    "documentation": 0.10,
    "communication_reflection": 0.10,
}

score_100 = sum(rubric_scores[k] * 20 * weighted[k] for k in rubric_scores)
print({"weighted_score": round(score_100, 2)})
```
