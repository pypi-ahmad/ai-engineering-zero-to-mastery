```python
from pathlib import Path

PROJECT_SKELETON = {
    "data": ["ingest.py", "prepare.py"],
    "model": ["train.py", "evaluate.py"],
    "serve": ["api.py"],
    "configs": ["config.yaml"],
    "docs": ["architecture.md", "report.md"],
}


def scaffold_project(root: str = "capstone_template"):
    root_path = Path(root)
    root_path.mkdir(parents=True, exist_ok=True)
    for folder, files in PROJECT_SKELETON.items():
        fp = root_path / folder
        fp.mkdir(parents=True, exist_ok=True)
        for f in files:
            p = fp / f
            if not p.exists():
                p.write_text(f"# TODO: implement {f}\n")
    return root_path

scaffold_project("tmp_capstone_template")
    
```

```python
data_ingest_template = "import pandas as pd\n\ndef load_data(path: str):\n    return pd.read_csv(path)\n"
train_template = "from sklearn.linear_model import LogisticRegression\n# TODO: train baseline and log metrics\n"
api_template = "from fastapi import FastAPI\napp = FastAPI()\n@app.get('/health')\ndef health(): return {'status':'ok'}\n"

print(data_ingest_template)
print(train_template)
print(api_template)
    
```
