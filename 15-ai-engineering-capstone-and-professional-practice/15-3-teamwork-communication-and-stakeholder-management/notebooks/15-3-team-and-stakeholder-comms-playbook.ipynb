## Scenario Exercises

### Scenario 1: Scope Creep
A mentor requests major feature additions two weeks before demo.

Response: clarify value, estimate effort/risk, reprioritize must/should/could, confirm in writing.

### Scenario 2: Team Priority Conflict
Disagreement between innovation and delivery stability.

Response: map options to success criteria, log decision, assign owner.

### Scenario 3: Unclear Requirements
Stakeholder asks to “make it smarter”.

Response: convert to measurable criteria and validate acceptance threshold.
    


## Capstone & Practice Case Studies & Exceptions

### Case Study A: Strong Tech, Weak Communication
Good model quality but weak decision records hurt final defense.

### Case Study B: Simpler Tech, Strong Communication
Clear updates and scope control led to better final assessment.

### Exception Pattern
Solo projects still require communication discipline with mentors/reviewers.
    


## Interview Questions & Answers

1. **Q:** How did you manage team coordination?  
   **A:** Weekly updates, role ownership, and decision logs.
2. **Q:** How did you handle conflicting priorities?

3. **Q:** What communication artifact was most useful?  
   **A:** Weekly update with blockers and action owners.
4. **Q:** How did you manage stakeholder expectations?  
   **A:** Clear acceptance criteria and regular cadence updates.
5. **Q:** How do you explain technical debt to non-technical stakeholders?

6. **Q:** How did you prevent scope creep?  
   **A:** Formal change review using must/should/could.
7. **Q:** Why keep meeting notes?  
   **A:** They preserve decisions and accountability.
8. **Q:** How did you resolve team conflict?  
   **A:** Criteria-based decisions and role clarity.
9. **Q:** Why include ethics in updates?

10. **Q:** What process change would you make next time?

11. **Q:** How does this help interviews?
