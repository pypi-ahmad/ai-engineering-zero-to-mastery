```python
import pandas as pd

reflection_prompts = [
    "What went better than expected?",
    "What did not go as planned?",
    "Which technical decision had biggest impact?",
    "Which communication decision changed outcomes?",
    "What would you change in version 2?",
    "How does this map to your target role?",
]

reflection_df = pd.DataFrame({"prompt": reflection_prompts, "your_notes": ["" for _ in reflection_prompts]})
reflection_df
    
```

## Portfolio Mapping Tool


```python
project_assets = pd.DataFrame(
    [
        {"asset": "Repository", "status": "draft", "improvement_needed": "clean setup instructions"},
        {"asset": "Architecture diagram", "status": "done", "improvement_needed": "add latency/cost notes"},
        {"asset": "Evaluation report", "status": "done", "improvement_needed": "segment analysis summary"},
        {"asset": "Demo video", "status": "draft", "improvement_needed": "record fallback path"},
        {"asset": "Interview STAR stories", "status": "draft", "improvement_needed": "add failure-recovery example"},
    ]
)

def readiness_score(df: pd.DataFrame) -> float:
    mapping = {"done": 1.0, "draft": 0.5, "missing": 0.0}
    return round(df["status"].map(mapping).mean() * 100, 2)

project_assets, readiness_score(project_assets)
    
```

## Interview Questions & Answers

1. **Q:** How did you structure your final capstone presentation?  
   **A:** Problem, approach, evidence, impact, limitations, and next steps.
2. **Q:** What is the strongest result you presented?  
   **A:** The KPI improvement tied directly to project objective.
3. **Q:** How did you explain limitations?  
   **A:** Explicitly, with mitigations and roadmap.
4. **Q:** What did you learn most from this capstone?  
   **A:** Balancing technical ambition with delivery reliability.
5. **Q:** What would you do differently next time?  
   **A:** Earlier instrumentation and tighter scope gates.
6. **Q:** How does this capstone support your career goals?  
   **A:** It demonstrates end-to-end ownership for target roles.
7. **Q:** How did you tailor your talk for non-technical audiences?  
   **A:** Focused on value and risk, not algorithm detail.
8. **Q:** What portfolio artifacts came from this project?  
   **A:** Repo, diagram, eval report, demo, and reflection notes.
9. **Q:** How did you handle Q&A on weak points?  
   **A:** Acknowledged limits and shared concrete next steps.
10. **Q:** How do you convert reflection into growth?  
    **A:** Translate lessons into explicit action items.
11. **Q:** What is your immediate post-capstone step?  
    **A:** One focused improvement sprint with measurable deliverables.
