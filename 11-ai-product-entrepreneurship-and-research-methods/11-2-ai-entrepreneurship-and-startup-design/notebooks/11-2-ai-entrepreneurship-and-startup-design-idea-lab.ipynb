## Idea Scoring Model

```python
@dataclass
class AIStartupIdea:
    name: str
    problem_severity: int      # 1-10
    ai_fit: int                # 1-10
    data_access: int           # 1-10
    moat_strength: int         # 1-10
    ethics_risk: int           # 1-10 (higher is worse)
    go_to_market_clarity: int  # 1-10


def score_idea(idea: AIStartupIdea) -> Dict[str, float]:
    # Weighted score with explicit ethics penalty.
    base = (
        0.24 * idea.problem_severity
        + 0.20 * idea.ai_fit
        + 0.18 * idea.data_access
        + 0.16 * idea.moat_strength
        + 0.12 * idea.go_to_market_clarity
    )
    penalty = 0.10 * idea.ethics_risk
    total = base - penalty
    return {
        "idea": idea.name,
        "base_score": round(base, 3),
        "ethics_penalty": round(penalty, 3),
        "total_score": round(total, 3),
    }


ideas: List[AIStartupIdea] = [
    AIStartupIdea("SMB Finance Copilot", 9, 8, 7, 7, 4, 8),
    AIStartupIdea("Generic Writing Bot", 5, 6, 5, 3, 5, 6),
    AIStartupIdea("Edge Predictive Maintenance", 8, 8, 6, 8, 3, 7),
    AIStartupIdea("Healthcare Triage Assistant", 9, 7, 5, 7, 8, 5),
]

scores_df = pd.DataFrame([score_idea(i) for i in ideas]).sort_values("total_score", ascending=False)
scores_df
```

## Interpret the Ranking

```python
scores_df.assign(
    recommendation=lambda df: df["total_score"].apply(
        lambda x: "Pilot" if x >= 6.5 else ("Validate More" if x >= 5 else "Deprioritize")
    )
)
```

## Business / Research Case Studies & Exceptions

### Case 1: AI copilot for SMB finance
- Strong pain and willingness-to-pay often make this a strong early-stage candidate.
- Must handle high trust requirements and auditability.

### Case 2: Edge predictive maintenance startup
- Good moat potential through deployment depth and sensor data loops.
- Risk: long enterprise sales cycles and integration burden.

### Case 3 (Exception): Generic horizontal assistant
- High competition, weak defensibility, and fragile retention.
- Better strategy may be a vertical specialization with measurable ROI.

## Interview Questions & Answers

1. **How do you identify AI startup opportunities?**  
Start from painful workflows, then test AI feasibility and willingness-to-pay.

2. **What should be in an AI Lean Canvas?**  
Problem, segment, value proposition, AI workflow, economics, risks, and moat.

3. **How do you evaluate AI fit?**  
Check whether AI materially outperforms deterministic alternatives.

4. **Why is ethics risk part of idea scoring?**  
High unresolved risk can block deployment and destroy trust.

5. **How do you think about moat in AI startups?**  
Workflow integration and proprietary feedback loops are stronger than model novelty alone.

6. **What is a common AI startup failure mode?**  
Building broad AI features without clear problem ownership.

7. **How do you choose pricing model for AI products?**  
Match pricing to value unit while protecting gross margin.

8. **When is usage pricing risky?**  
When inference cost volatility outpaces revenue per task.

9. **How do you de-risk GTM for AI ventures?**  
Pilot with one segment and prove a narrow value thesis first.

10. **How do you integrate responsible AI in a startup culture?**  
Use pre-launch risk checks, decision logs, and clear escalation rules.

11. **How do you decide whether to pivot?**  
If repeated experiments show weak retention or poor unit economics despite iteration.
