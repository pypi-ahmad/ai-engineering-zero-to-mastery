# Lesson 11.4 - Portfolio, Capstone & Publication / Launch (planner notebook)

This notebook helps you convert scattered projects into a coherent AI portfolio strategy with quantitative gap analysis.

## Objectives

1. Inventory existing projects.
2. Score projects on technical, product, domain, and research dimensions.
3. Identify gaps and generate next capstone ideas.

```python
from __future__ import annotations

from dataclasses import dataclass
from typing import List

import pandas as pd
```

## Portfolio Inventory Template

For each project, capture:

- problem statement,
- users/stakeholders,
- data and model stack,
- deployment or demo status,
- evaluation evidence,
- ethical/risk controls,
- measurable impact.

```python
@dataclass
class ProjectProfile:
    name: str
    technical_depth: int
    product_thinking: int
    domain_relevance: int
    research_rigor: int
    communication_quality: int


def score_project(p: ProjectProfile) -> float:
    return round(
        0.30 * p.technical_depth
        + 0.25 * p.product_thinking
        + 0.15 * p.domain_relevance
        + 0.20 * p.research_rigor
        + 0.10 * p.communication_quality,
        2,
    )


projects: List[ProjectProfile] = [
    ProjectProfile("RAG Support Assistant", 8, 8, 7, 6, 7),
    ProjectProfile("Credit Risk ML Pipeline", 7, 7, 8, 7, 6),
    ProjectProfile("CV Defect Detection Demo", 7, 5, 6, 5, 5),
    ProjectProfile("Agentic Workflow Prototype", 8, 7, 6, 6, 6),
]

portfolio_df = pd.DataFrame(
    [
        {
            "project": p.name,
            "technical_depth": p.technical_depth,
            "product_thinking": p.product_thinking,
            "domain_relevance": p.domain_relevance,
            "research_rigor": p.research_rigor,
            "communication_quality": p.communication_quality,
            "score": score_project(p),
        }
        for p in projects
    ]
).sort_values("score", ascending=False)

portfolio_df
```

## Gap Analysis and Next Capstone Suggestions

```python
dimension_means = portfolio_df[
    [
        "technical_depth",
        "product_thinking",
        "domain_relevance",
        "research_rigor",
        "communication_quality",
    ]
].mean().sort_values()

dimension_means
```

```python
def suggest_next_capstones(dimension_scores: pd.Series) -> list[str]:
    suggestions = []
    weakest = dimension_scores.index[0]
    if weakest == "research_rigor":
        suggestions.append("Build a research-style capstone with baselines, ablations, and reproducibility package.")
    if weakest == "product_thinking":
        suggestions.append("Build an AI feature from PRD to MVP with metric tree and A/B evaluation plan.")
    if weakest == "communication_quality":
        suggestions.append("Publish one full project report plus demo walkthrough video.")
    if weakest == "domain_relevance":
        suggestions.append("Ship a domain-specific capstone in one target industry (health, finance, or robotics).")
    if weakest == "technical_depth":
        suggestions.append("Deliver a production-grade system with monitoring, rollback, and reliability tests.")
    return suggestions


suggest_next_capstones(dimension_means)
```

## Interview Questions & Answers

1. **Q: How do you choose which projects to highlight?**
   **A:** Pick projects that show end-to-end ownership and measurable outcomes.

2. **Q: What makes a capstone portfolio-ready?**
   **A:** Technical rigor, impact evidence, clear documentation, and trade-off discussion.

3. **Q: How do you explain business impact from technical work?**
   **A:** Map model/system metrics to user outcomes and economic effects.

4. **Q: How do you present a failed experiment in interviews?**
   **A:** Describe hypothesis, failure signal, root-cause analysis, and design change.

5. **Q: How many projects should a portfolio include?**
   **A:** Fewer but deeper projects are usually stronger than many shallow demos.

6. **Q: How do you adapt your portfolio for PM vs engineering roles?**
   **A:** PM: outcomes and decisions; engineering: architecture, reliability, and performance.

7. **Q: What does publication-ready documentation include?**
   **A:** Methodology, reproducibility details, limitations, and evidence-backed conclusions.

8. **Q: How do you identify portfolio gaps?**
   **A:** Score projects across dimensions and improve weakest dimensions with next capstone.

9. **Q: How do you validate impact before full launch?**
   **A:** Run pilots with explicit success criteria and baseline comparisons.

10. **Q: How do you discuss responsible AI work in interviews?**
   **A:** Show practical controls: risk reviews, monitoring, and incident response decisions.

11. **Q: What should you do after finishing a capstone?**
   **A:** Iterate based on feedback, harden reliability, and publish an updated roadmap.
